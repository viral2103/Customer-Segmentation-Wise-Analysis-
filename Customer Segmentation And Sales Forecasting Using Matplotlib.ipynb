```python
df=pd.read_csv('superstore.csv',encoding='ISO-8859-1')
df.index=range(1,len(df) + 1)
df
```

<b>1)How do total sales and profit vary over each month or year?</b>

```python
import pandas as pd
import matplotlib.pyplot as plt

df['Order Date'] = pd.to_datetime(df['Order Date'])
df.set_index('Order Date', inplace=True)

trends=df.resample('YE').sum()

plt.figure(figsize=(14, 6))
plt.plot(trends.index, trends['Sales'], label='Sales', marker='o')
plt.plot(trends.index, trends['Profit'], label='Profit', marker='o')
plt.title('Profit and Sales Yearwise')
plt.xlabel('Year')
plt.ylabel('Amount')
plt.legend()
plt.grid(True)
# plt.tight_layout()
plt.show()

```

<b>2)Are there any seasonal patterns in sales across the years?</b>

```python
# Convert 'Order Date' to datetime
df['Order Date'] = pd.to_datetime(df['Order Date'])

# Extract year and month
df['Year'] = df['Order Date'].dt.year
df['Month'] = df['Order Date'].dt.month_name()

# Optional: Order months properly
month_order = ['January', 'February', 'March', 'April', 'May', 'June',
               'July', 'August', 'September', 'October', 'November', 'December']

df['Month'] = pd.Categorical(df['Month'], categories=month_order, ordered=True)

# Group sales by Year and Month
monthly_sales = df.groupby(['Year', 'Month'])['Sales'].sum().unstack(level=0)

# Plotting
plt.figure(figsize=(14, 6))
monthly_sales.plot(kind='line', marker='o', figsize=(14,6))
plt.title('Seasonal Sales Trends Across Years')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.grid(True)
plt.legend(title='Year')
plt.tight_layout()
plt.show()
```

<b>3)How does the number of orders trend over time?</b>

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])

# Group by month and count unique orders
order_trend = df.groupby(pd.Grouper(key='Order Date', freq='ME'))['Order ID'].nunique()
# Plotting
plt.figure(figsize=(14, 6))
order_trend.plot(kind='line', marker='o', color='purple')
plt.title('Number of Orders Over Time')
plt.xlabel('Order Date')
plt.ylabel('Number of Orders')
plt.grid(True)
plt.tight_layout()
plt.show()
```

<b>20)What is the trend of profit margin over time?</b>

```python
# Ensure 'Order Date' is datetime
df['Order Date'] = pd.to_datetime(df['Order Date'])

# Filter out zero sales to avoid division by zero
df['Profit Margin'] = df['Profit'] / df['Sales']

# Set datetime index
df.set_index('Order Date', inplace=True)

# Resample monthly and calculate average profit margin
monthly_profit_margin = df['Profit Margin'].resample('ME').mean()

# Plot
plt.figure(figsize=(19, 8))
monthly_profit_margin.plot(marker='o', color='teal')
plt.title('Profit Margin Trend Over Time')
plt.xlabel('Month')
plt.ylabel('Average Profit Margin')
plt.xticks(rotation=0)
plt.grid(True)
plt.tight_layout()
plt.show()
```
