<b>18)dentify customers who generated the highest sales Month Wise</b>

```python
# Ensure Order Date is datetime
df['Order Date'] = pd.to_datetime(df['Order Date'])

# Set DateTime index
df.set_index('Order Date', inplace=True)

# Group by Customer and resample monthly
top_customers = df.groupby('Customer Name').resample('ME')[['Quantity', 'Sales']].sum()

# Sort by Sales in descending order
top_customers = top_customers.sort_values(by='Sales', ascending=False)

top_customers.to_csv('Customer Purchase Most Product With Highest Sales.csv')
top_customers
```
